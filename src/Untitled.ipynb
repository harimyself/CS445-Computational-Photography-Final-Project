```python
import cv2
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

def readFrames(video_file_name):
    vc = cv2.VideoCapture(video_file_name)

    if vc.isOpened():
        rval, frame = vc.read()
    else:
        rval = False

    frames = []
    
    while rval:
        rval, frame = vc.read()
        frames.append(frame)
    
    
    vc.release()
        
    return frames
```

```python
def convertToGreyScaleAndDetectKP(image, sift):
    image_grey = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

    kp, des = sift.detectAndCompute(image_grey, None)
    
    return kp, des

def matchDes(template_des, frame_des):
    bf = cv2.BFMatcher()
    matches = bf.knnMatch(template_des, frame_des, k=2)
    
    fair_matches = []
    for m,n in matches:
        if m.distance < 0.75*n.distance:
            fair_matches.append([m])
        
    return fair_matches, len(fair_matches)/len(template_des)

def stopSignMatcher(frame_ip, sift):
    frame_kp, frame_des = convertToGreyScaleAndDetectKP(frame_ip, sift)
    fair_maches, template_mpercntage = matchDes(template_des, frame_des)
    
    return len(fair_maches) > 3

def getStopSignDescriptor(sift):
    traffic_signs_dir = '/Users/hbojja/uiuc/CS445-CP/FinalProject/input/traffic_signs/'
    template = cv2.imread(traffic_signs_dir+ 'stop_sign_4.jpeg')
    template_kp, template_des = convertToGreyScaleAndDetectKP(template, sift)
    
    return template_kp, template_des
```

```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, push_notebook
import cv2
import time



input_video_file = '/Users/hbojja/uiuc/CS445-CP/FinalProject/input/tape_1.mov'

cap = cv2.VideoCapture(input_video_file)
ret, frame = cap.read()
frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA) # because Bokeh expects a RGBA image
frame=cv2.flip(frame, -1) # because Bokeh flips vertically
width=frame.shape[1]
height=frame.shape[0]
p = figure(x_range=(0,width), y_range=(0,height), output_backend="webgl", width=width, height=height)
myImage = p.image_rgba(image=[frame], x=0, y=0, dw=width, dh=height)
show(p, notebook_handle=True)

while True:
    ret, frame = cap.read()
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGBA)
    frame=cv2.flip(frame, -1)
    myImage.data_source.data['image']=[frame]
    push_notebook()
    time.sleep(0.3)
        
cap.release()
   
# Closes all the frames
cv2.destroyAllWindows()
```
