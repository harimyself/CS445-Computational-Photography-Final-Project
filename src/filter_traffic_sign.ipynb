### Filter signals frames

```python
import cv2
import datetime

input_video_path = '/Users/hbojja/uiuc/CS445-CP/FinalProject/input/collection/3.mov'

train_data_path = '/Users/hbojja/uiuc/CS445-CP/FinalProject/train_data/'

cap = cv2.VideoCapture(input_video_path)
```

```python
frame_pointer = 0
frame_processed = 0

stop_count = 5000
school_zone_count = 5000
speed_limit_25_count = 5000
speed_limit_35_count = 5000
cross_walk_count = 5000

print(datetime.datetime.now())

while (True):
    try:

        frame_processed = frame_processed + 1
        
        ret, frame = cap.read()
            
        frame_pointer = frame_pointer + 15
        
        cap.set(1, frame_pointer)
        
        gray = cv2.cvtColor(frame.astype('uint8'), cv2.COLOR_BGR2GRAY)

        rec_label, fair_des_count = traffic_sign_recogninzer_filter.realtime_template_matcher(gray)

        print(rec_label, fair_des_count)
        if rec_label == 'stop' :
            stop_count = stop_count + 1
            cv2.imwrite(train_data_path + 'stop/' +"stop_%d.jpg" % stop_count, frame)

        elif rec_label == 'school_zone' :
            school_zone_count = school_zone_count + 1
            cv2.imwrite(train_data_path + 'school_zone/' + "school_zone_%d.jpg" % school_zone_count, frame)

        elif rec_label == 'speed_limit_25' :
            speed_limit_25_count = speed_limit_25_count + 1
            cv2.imwrite(train_data_path + 'speed_limit_25/' + "speed_limit_25_%d.jpg" % speed_limit_25_count, frame)

        elif rec_label == 'speed_limit_35' :
            speed_limit_35_count = speed_limit_35_count + 1
            cv2.imwrite(train_data_path + 'speed_limit_35/' + "speed_limit_35_%d.jpg" % speed_limit_35_count, frame)

        elif rec_label == 'cross_walk' :
            cross_walk_count = cross_walk_count + 1
            cv2.imwrite(train_data_path + 'cross_walk/' + "cross_walk_%d.jpg" % cross_walk_count, frame)

        elif rec_label == '':
            cv2.imwrite(train_data_path + 'other/' + "other_%d.jpg" % cross_walk_count, frame)
        
    except Exception as e: print(e)
        
        

print("Frames processed:", frame_processed, frame_pointer)
print(datetime.datetime.now())

cap.release()
cv2.destroyAllWindows()
        
        
        
```
