```python
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

input_dir = '/Users/hbojja/uiuc/CS445-CP/FinalProject/train_data_individual/other/'

output_dir = '/Users/hbojja/uiuc/CS445-CP/FinalProject/train_data_individual/cleansed/other/'

for filename in os.listdir(input_dir):
    if filename == '.DS_Store':
        continue
        
    print(input_dir + filename)
    
    image = cv2.imread(input_dir + filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print(image.shape)
    
    scale_percent = 50 # percent of original size
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
  
    # resize image
    resized = cv2.resize(image, dim)
 
    print('Resized Dimensions : ', resized.shape)
    
    resized = resized[71:-60, int(width/2):]
    
    print('Resized Dimensions : ', resized.shape)
    
    cv2.imwrite(output_dir + filename, resized)
#     plt.figure()
#     plt.imshow(resized)

#     break
```
