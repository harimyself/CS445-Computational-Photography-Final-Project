```python


input_video_path = '/Users/hbojja/uiuc/CS445-CP/FinalProject/input/collection/4.mov'
# input_video_path = '/Users/hbojja/uiuc/CS445-CP/FinalProject/input/tape_2.mov'


cap = cv2.VideoCapture(input_video_path)
```

```python
frame_pointer = 0
frame_processed = 422

stop_count = 0
school_zone_count = 0
speed_limit_25_count = 0
speed_limit_35_count = 0
cross_walk_count = 0

print(datetime.datetime.now())

while (True):
    try:
        frame_processed = frame_processed + 1
        
        ret, frame = cap.read()
        
        if ret:
            
            frame_pointer = frame_pointer + 15

            cap.set(1, frame_pointer)

            gray = cv2.cvtColor(frame.astype('uint8'), cv2.COLOR_BGR2GRAY)
            sign_iden = traffic_sign_recogninzer_filter_individual.realtime_template_matcher(gray)
            
            if sign_iden == True:
                print(train_data_path + "/frame_%d.jpg" % frame_processed)
                cv2.imwrite(train_data_path + "/frame_%d.jpg" % frame_processed, frame)
        else:
            cap.release()
            break
        
    except Exception as e: print(e)
        
        

print("Frames processed:", frame_processed, frame_pointer)
print(datetime.datetime.now())

cv2.destroyAllWindows()
        
        
        
```
